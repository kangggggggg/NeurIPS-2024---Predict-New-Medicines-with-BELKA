```python
import numpy as np

# 각 단백질별로 0과 1의 비율이 동일하도록 샘플링
def balanced_sampling(df, col_name, sample_frac=0.005, random_state=42):
    df_0 = df[df[col_name] == 0].sample(frac=sample_frac, random_state=random_state)
    df_1 = df[df[col_name] == 1].sample(frac=sample_frac, random_state=random_state)
    return pd.concat([df_0, df_1])


percent = 0.01/2
sampled_BRD4 = balanced_sampling(train, 'binds_BRD4', sample_frac=percent/3)
sampled_HSA = balanced_sampling(train, 'binds_HSA', sample_frac=percent/3)
sampled_sEH = balanced_sampling(train, 'binds_sEH', sample_frac=percent/3)

train = pd.concat([sampled_BRD4, sampled_HSA, sampled_sEH])
sampled_BRD4 = 0
sampled_HSA = 0
sampled_sEH = 0

train.shape
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import average_precision_score
from sklearn.preprocessing import OneHotEncoder

# Convert SMILES to RDKit molecules
train['molecule'] = train['molecule_smiles'].apply(Chem.MolFromSmiles)

# Generate ECFPs
def generate_ecfp(molecule, radius=2, bits=1024):
    if molecule is None:
        return None
    return list(AllChem.GetMorganFingerprintAsBitVect(molecule, radius, nBits=bits))

train['ecfp'] = train['molecule'].apply(generate_ecfp)

train.head()
```

```python
train.drop(columns=['buildingblock1_smiles', 'buildingblock2_smiles',
                    'buildingblock3_smiles', 'molecule_smiles', 'molecule'])
```

```python
def rf(protein_name, test_size=0.2):
  X = train['ecfp'].tolist()
  y = train['binds_'+protein_name].tolist()

  # Split the data into train and test sets
  X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=test_size, random_state=42)
  model=RandomForestClassifier(n_estimators=100, random_state=42)
  model.fit(X_train, y_train)

  # Make predictions on the test set
  y_pred_proba = model.predict_proba(X_val)[:, 1]  # Probability of the positive class

  # Calculate the mean average precision
  map_score = average_precision_score(y_val, y_pred_proba)
  print(f"Mean Average Precision (mAP): {map_score:.2f}")

  return model, map_score
```

```python
model_BRD4, mAP_BRD4 = rf('BRD4')
```

```python
model_HSA, mAP_HSA = rf('HSA')
```

```python
model_sEH, mAP_sEH = rf('sEH')
```

```python
# 각 프로틴에 대한 모델 설정
models = {
    'BRD4': model_BRD4,
    'HSA': model_HSA,
    'sEH': model_sEH
}

# 각 모델의 mAP (Mean Average Precision)
model_mAPs = {
    'BRD4': mAP_BRD4,
    'HSA': mAP_HSA,
    'sEH': mAP_sEH
}

# 테스트 데이터 파일 경로와 출력 파일 경로
test_file = DIR_PATH + 'test.csv'
submission = pd.read_csv('/content/leash-BELKA/sample_submission.csv')
binds_list = []

df_test = pd.read_csv(test_file)

# molecule_smiles로부터 ECFP 생성
df_test['molecule'] = df_test['molecule_smiles'].apply(Chem.MolFromSmiles)
df_test['ecfp'] = df_test['molecule'].apply(generate_ecfp)

X_test = df_test['ecfp'].tolist()

for i, row in df_test.iterrows():
    probabilities = []
    weights = []
    for protein in ['BRD4', 'HSA', 'sEH']:
        if row[f'is_{protein}']:
            model = models[protein]
            prob = model.predict_proba([X_test[i]])[0, 1]
            probabilities.append(prob)
            weights.append(model_mAPs[protein])

    if probabilities:
        weighted_average = sum(p * w for p, w in zip(probabilities, weights)) / sum(weights)
    else:
        weighted_average = None
    binds_list.append(weighted_average)

submission['binds'] = binds_list
submission.to_csv('submission.csv', index=False)
```
